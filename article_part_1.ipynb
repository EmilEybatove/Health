Возьмём константу, появившуюся здесь магически образом: список всех необходимых колонок из полного датасета

объединим составляющие cesd в целое и удалим их

```python
cesd1 <- full_data[, c("r1deprex", "r1efforx", "r1sleepx", "r1whappx", "r1flonex",  "r1fsadx", "r1goingx",  "r1enlifx")]
cesd1$r1whappx <- -1 * cesd1$r1whappx
cesd1$r1enlifx <- -1 * cesd1$r1enlifx
```

### возможно был реверсированый score. нужно подумать. но потом

```python
full_data$r1cesd <- (22 - rowSums(cesd1)) / 3
```
